## 2 - Criação dos dataframes e análise dos dados

Sequência de procedimentos iniciais realizados:

- Em seguida, foi realizada a mudança de codificação desses arquivos para o padrão: UTF-8, pois houve dificuldade na manipulação dos arquivos no padrão de codificação original utilizado nos arquivos disponibilizados no sítio da PRF.

A seguir, seguem os passo para manipulação dos datasets e dataframes utilizados.

```python
# Importando as bibliotecas utilizadas no projeto

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

## Configurando o pandas para exibição de um número alto de linhas
pd.set_option('max_rows', 99999)
```

```python
# Mapeando URLs dos datasets dos acidentes de trânsito (2010 a 2021)

acidentes2010URL = "https://raw.githubusercontent.com/allexsantosrn/Ciencia-de-Dados/main/files/datatran2010.csv"
acidentes2011URL = "https://raw.githubusercontent.com/allexsantosrn/Ciencia-de-Dados/main/files/datatran2011.csv"
acidentes2012URL = "https://raw.githubusercontent.com/allexsantosrn/Ciencia-de-Dados/main/files/datatran2012.csv"
acidentes2013URL = "https://raw.githubusercontent.com/allexsantosrn/Ciencia-de-Dados/main/files/datatran2013.csv"
acidentes2014URL = "https://raw.githubusercontent.com/allexsantosrn/Ciencia-de-Dados/main/files/datatran2014.csv"
acidentes2015URL = "https://raw.githubusercontent.com/allexsantosrn/Ciencia-de-Dados/main/files/datatran2015.csv"
acidentes2016URL = "https://raw.githubusercontent.com/allexsantosrn/Ciencia-de-Dados/main/files/datatran2016.csv"
acidentes2017URL = "https://raw.githubusercontent.com/allexsantosrn/Ciencia-de-Dados/main/files/datatran2017.csv"
acidentes2018URL = "https://raw.githubusercontent.com/allexsantosrn/Ciencia-de-Dados/main/files/datatran2018.csv"
acidentes2019URL = "https://raw.githubusercontent.com/allexsantosrn/Ciencia-de-Dados/main/files/datatran2019.csv"
acidentes2020URL = "https://raw.githubusercontent.com/allexsantosrn/Ciencia-de-Dados/main/files/datatran2020.csv"
acidentes2021URL = "https://raw.githubusercontent.com/allexsantosrn/Ciencia-de-Dados/main/files/datatran2021.csv"
```

```python
# Criando dataframes dos acidentes de trânsito por ano

acidentes2010 = pd.read_csv(acidentes2010URL, sep=';')
acidentes2011 = pd.read_csv(acidentes2011URL, sep=';')
acidentes2012 = pd.read_csv(acidentes2012URL, sep=';', low_memory=False)
acidentes2013 = pd.read_csv(acidentes2013URL, sep=';')
acidentes2014 = pd.read_csv(acidentes2014URL, sep=';')
acidentes2015 = pd.read_csv(acidentes2015URL, sep=';')
acidentes2016 = pd.read_csv(acidentes2016URL, sep=';')
acidentes2017 = pd.read_csv(acidentes2017URL, sep=';')
acidentes2018 = pd.read_csv(acidentes2018URL, sep=';')
acidentes2019 = pd.read_csv(acidentes2019URL, sep=';')
acidentes2020 = pd.read_csv(acidentes2020URL, sep=';')
acidentes2021 = pd.read_csv(acidentes2021URL, sep=';')
```

```python
# Mapeando apenas os acidentes de trânsito no estado do Rio Grande do Norte

acidentesRN2010 = acidentes2010.query('uf == "RN"')
acidentesRN2011 = acidentes2011.query('uf == "RN"')
acidentesRN2012 = acidentes2012.query('uf == "RN"')
acidentesRN2013 = acidentes2013.query('uf == "RN"')
acidentesRN2014 = acidentes2014.query('uf == "RN"')
acidentesRN2015 = acidentes2015.query('uf == "RN"')
acidentesRN2016 = acidentes2016.query('uf == "RN"')
acidentesRN2017 = acidentes2017.query('uf == "RN"')
acidentesRN2018 = acidentes2018.query('uf == "RN"')
acidentesRN2019 = acidentes2019.query('uf == "RN"')
acidentesRN2020 = acidentes2020.query('uf == "RN"')
acidentesRN2021 = acidentes2021.query('uf == "RN"')
```

```python
# Concatenando os dataframes dos acidentes de trânsito no RN

acidentesRN = pd.concat([acidentesRN2010, acidentesRN2011, acidentesRN2012,
                         acidentesRN2013, acidentesRN2014, acidentesRN2015,
                         acidentesRN2016, acidentesRN2017, acidentesRN2018,
                         acidentesRN2019, acidentesRN2020, acidentesRN2021])

# Concatenando todos os acidentes de trânsito no Brasil

acidentesBrasil = pd.concat([acidentes2010,acidentes2011,acidentes2012,
                             acidentes2013,acidentes2014,acidentes2015,
                             acidentes2016,acidentes2017,acidentes2018,
                             acidentes2019,acidentes2020,acidentes2021])

# Tranformando coluna da data do acidente no tipo datetime
acidentesBrasil['data_inversa'] = pd.to_datetime(acidentesBrasil["data_inversa"])
```

## 3 - Plotagem dos gráficos e análise dos dados

A partir dos dataframes criados, as seguintes análises foram realizadas:

```python
# Gerando gráfico do total de acidentes por ano no Brasil

df = pd.DataFrame({'Total': [len(acidentes2010), len(acidentes2011), len(acidentes2012), 
                             len(acidentes2013), len(acidentes2014), len(acidentes2015), 
                             len(acidentes2016), len(acidentes2017), len(acidentes2018), 
                             len(acidentes2019), len(acidentes2020), len(acidentes2021)], 
                   'Ano': [2010, 2011, 2012, 2013, 2014, 2015, 
                           2016, 2017, 2018, 2019, 2020, 2021]})

fig = sns.barplot(y="Total", x="Ano", data=df, color="salmon")
plt.title("Quantitativo de acidentes por ano no Brasil.")
plt.show(fig)

fig2 = sns.lineplot(y="Total", x="Ano", data=df)
plt.title("Quantitativo de acidentes por ano no Brasil.")
plt.show(fig2)
```

Este primeiro gráfico que retrata o panorama nacional dos acidentes de trânsito é bastante interessante, pois apresenta uma boa notícia: O número de acidentes de trânsito nas rodovias federais vem diminuindo ao longo dos anos no Brasil. Saímos de um valor total de 180 mil acidentes por ano em 2010, para um total de aproximadamente 60 mil acidentes por ano em 2021.

```python
# Gerando gráfico do total de acidentes por ano no estado do Rio Grande do Norte

df = pd.DataFrame({'Total': [len(acidentesRN2010), len(acidentesRN2011), len(acidentesRN2012), 
                             len(acidentesRN2013), len(acidentesRN2014), len(acidentesRN2015), 
                             len(acidentesRN2016), len(acidentesRN2017), len(acidentesRN2018), 
                             len(acidentesRN2019), len(acidentesRN2020), len(acidentesRN2021)],
                    
                   'Ano': [2010, 2011, 2012, 2013, 2014, 2015, 
                           2016, 2017, 2018, 2019, 2020, 2021]})

fig = sns.barplot(y="Total", x="Ano", data=df, color="salmon")
plt.title("Quantitativo de acidentes por ano no Rio Grande do Norte.")
plt.show(fig)

fig2 = sns.lineplot(y="Total", x="Ano", data=df)
plt.title("Quantitativo de acidentes por ano no Rio Grande do Norte.")
plt.show(fig2)
```

O estado do Rio Grande do Norte acompanha a média brasileira de diminuição de acidentes em suas rodovias federais, saindo de um valor de aproximadamente 4 mil acidentes por ano em 2010 para menos de 1500 acidentes em 2021.

```python
# Gerando gráfico com o total de acidentes por causa/motivo no RN

## Mapeando as causas de acidentes mais frequentes no RN
#acidentesRN['causa_acidente'].value_counts(ascending=False).head(12)

atencao = acidentesRN.query('(causa_acidente == "Falta de atenção") or (causa_acidente == "Falta de Atenção à Condução") ')
outras = acidentesRN.query('causa_acidente == "Outras" ')
distancia = acidentesRN.query('causa_acidente == "Não guardar distância de segurança" ')
animais = acidentesRN.query('causa_acidente == "Animais na Pista" ')
sinalizacao = acidentesRN.query('causa_acidente == "Desobediência à sinalização" ')
alcool = acidentesRN.query('(causa_acidente == "Ingestão de álcool") or (causa_acidente == "Ingestão de Álcool") ')
normas = acidentesRN.query('causa_acidente == "Desobediência às normas de trânsito pelo condutor" ')
defeito = acidentesRN.query('causa_acidente == "Defeito mecânico em veículo" ')
ultrapassagem = acidentesRN.query('causa_acidente == "Ultrapassagem indevida" ')
dormindo = acidentesRN.query('causa_acidente == "Dormindo" ')

df = pd.DataFrame({'Total': [len(atencao), len(outras), len(distancia), len(animais), 
                            len(sinalizacao), len(alcool), len(normas), len(defeito),
                            len(ultrapassagem), len(dormindo)], 

                   'Motivo': ["Falta de Atenção à Condução", "Outras", "Não guardar distância de segurança", 
                              "Animais na Pista", "Desobediência à sinalização", "Ingestão de Álcool", 
                              "Desobediência às normas de trânsito pelo condutor", "Defeito mecânico em veículo",
                              "Ultrapassagem indevida", "Dormindo"]})

fig = sns.barplot(y="Motivo", x="Total", data=df)
plt.title("Total de acidentes por causa/motivo no estado do RN.")
plt.show(fig)
```

Conforme identificado pelo gráfico, os tipos de acidentes mais comuns são os que envolvem colisões, em principal, colisões traseiras.

```python
# Gerando gráfico do total de acidentes por mês no Brasil

acidentesBrasilTotalPorMes = []

for i in range(1,13):
    acidentesBrasilTotalPorMes.append(len(acidentesBrasil.loc[acidentesBrasil['data_inversa'].dt.month == i]))


dfMesBrasil = pd.DataFrame({'Total de Acidentes': acidentesBrasilTotalPorMes, 
                   'Mês': ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 
                           'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']})

figMesBrasil = sns.barplot(y="Total de Acidentes", x="Mês", data=dfMesBrasil, color="salmon")
plt.title("Quantitativo de acidentes por mês no Brasil entre 2010 e 2021")
plt.show(figMesBrasil)


# Gerando gráfico do total de acidentes por mês no RN
acidentesRNTotalPorMes = []

for i in range(1,13):
    acidentesRNTotalPorMes.append(len(acidentesRN.loc[pd.to_datetime(acidentesRN['data_inversa']).dt.month == i]))


dfMesRn = pd.DataFrame({'Total de Acidentes': acidentesRNTotalPorMes, 
                   'Mês': ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 
                           'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']})

figMesRN = sns.barplot(y="Total de Acidentes", x="Mês", data=dfMesRn, color="blue")
plt.title("Quantitativo de acidentes por mês no RN entre 2010 e 2021")
plt.show(figMesRN)
```

Os gráficos abaixo mostram o total de acidentes nesses anos em relação aos meses em que ocorreram, é interessante notar que apesar da diferença entre os meses não ser grande, os meses que mais tem acidentes são Janeiro e Dezembro, provavelmente pelo aumento de viagens devido aos feriados e as férias.

```python
# Gerando gráfico com o total de acidentes por dia da semana no RN

## Mapeando o total de acidentes por dia da semana
#acidentesRN['dia_semana'].value_counts(ascending=False)

segunda = acidentesRN.query('(dia_semana == "Segunda") or (dia_semana == "segunda-feira") ')
terça = acidentesRN.query('(dia_semana == "Terça") or (dia_semana == "terça-feira") ')
quarta = acidentesRN.query('(dia_semana == "Quarta") or (dia_semana == "quarta-feira") ')
quinta = acidentesRN.query('(dia_semana== "Quinta") or (dia_semana == "quinta-feira") ')
sexta = acidentesRN.query('(dia_semana == "Sexta") or (dia_semana == "sexta-feira") ')
sabado = acidentesRN.query('(dia_semana == "Sábado") or (dia_semana == "sábado") ')
domingo = acidentesRN.query('(dia_semana == "Domingo") or (dia_semana == "domingo") ')

df = pd.DataFrame({'Total': [len(segunda), len(terça), len(quarta), len(quinta), 
                            len(sexta), len(sabado), len(domingo)], 

                   'Dia': ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", 
                            "Sábado", "Domingo"]})

fig = sns.barplot(y="Dia", x="Total", data=df)
plt.title("Total de acidentes por dia da semana no RN.")
plt.show(fig)
```

O gráfico mostra que a maior parte dos acidentes ocorre na sexta-feira e sábado. Em geral são dias em que as pessoas viajam ou saem a noite. É possível também uma correlação com o uso de bebida alcóolica que costuma ocorrer nos fins de semana.
